```
#-= Subtraction 
    ( a -= 1 ) ( a = a - 1 )
 
```

```
 # *= Multiplication 
    ( a *= 1 ) ( a = a * 1 )
 
```

```
 # /= Division 
    ( a /= 1 ) ( a = a / 1 )
 
```

```
 # %= Modulus 
    ( a %= 1 ) ( a = a % 1 )
 
```

```
#  **= Exponent 
    ( a **= 1 ) ( a = a ** 1 )
  
```

```
# //= Integer Division or Floor Division 
    ( a //= 1 ) ( a = a // 1 )
```
